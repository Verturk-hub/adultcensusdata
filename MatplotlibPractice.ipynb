```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import io
df = pd.read_csv(io.BytesIO(uploaded['company_sales_data.csv']))
df.head()
```

```python
import matplotlib.pyplot as plt
monthlist = df['month_number'].tolist()
profitlist = df['total_profit'].tolist()
plt.plot(monthlist, profitlist, color='r', marker='o', linewidth=3, linestyle='--')
#plt.xlabel('Month Number')
#plt.ylabel('Total Profit')
plt.xticks(monthlist)
plt.yticks([100000,200000,300000,400000,500000])
plt.title('Profit data of last year')
plt.legend('profit', loc='lower right')
plt.show()
```

```python
bssales = df['bathingsoap'].tolist()
sssales = df['shampoo'].tolist()
fwsales = df['facewash'].tolist()
plt.plot(monthlist, bssales, label = 'bathing soap')
plt.plot(monthlist, sssales, label = 'shampoo')
plt.plot(monthlist,fwsales, label = 'face wash')
plt.legend(loc='upper left')
```

```python
tp = df['toothpaste'].tolist()
ml = df['month_number']. tolist()
plt.scatter(ml,tp, label='Toothpaste Sale')
plt.legend()
plt.grid(True, linewidth=1, linestyle='--')
plt.xticks(monthlist)
plt.show()
```

```python
fwsales = df['facewash'].tolist()
fcsales = df['facecream'].tolist()
plt.bar([a-0.25 for a in monthlist],fcsales, width=0.25, label='Face Cream Sales', align='edge')
plt.bar([a+0.25 for a in monthlist],fwsales, width=0.25, label='Face Wash Sales', align='edge')
plt.legend()
plt.xticks(monthlist)
plt.show()
```

```python
salesData = [df['bathingsoap'].sum(), df['shampoo'].sum(), df['facewash'].sum(), df['facecream'].sum(), df['toothpaste'].sum(), df['moisturizer'].sum()]
labels = ['Bathing soap','Shampoo','FaceWash', 'FaceCream', 'ToothPaste', 'Moisturizer']
plt.pie(salesData, labels=labels, autopct='%1.1f%%')
```

```python
bathingsoap   = df ['bathingsoap'].tolist()
faceWashSalesData   = df ['facewash'].tolist()

f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(monthlist, bathingsoap, label = 'Bathingsoap Sales Data', color='k', marker='o', linewidth=3)
axarr[0].set_title('Sales Data of a Bathingsoap')
axarr[1].plot(monthlist, faceWashSalesData, label = 'Face Wash Sales Data', color='r', marker='o', linewidth=3)
axarr[1].set_title('Sales Data of a Facewash')

plt.xticks(monthlist)
#plt.xlabel('Month Number')
#plt.ylabel('Sales units in number')
plt.show()
```
